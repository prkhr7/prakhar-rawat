```r
data(iris)
head(iris)
```

```r
par(mfrow = c(2, 2))
hist(iris$Sepal.Length, breaks = 20)
hist(iris$Sepal.Width, breaks = 20)
```

```r
hist(iris$Petal.Length, breaks = 20)
hist(iris$Petal.Width, breaks = 20)
```

```r
log.iris <- log(iris[, 1:4]) #log transform 
iris.species <- iris[, 5]
par(mfrow = c(2, 2))
hist(log.iris$Sepal.Length, breaks = 20)
hist(log.iris$Sepal.Width, breaks = 20)
hist(log.iris$Petal.Length, breaks = 20)
hist(log.iris$Petal.Width, breaks = 20)
```

```r
ir.pca <- prcomp(log.iris, center = TRUE, scale = TRUE) #applying PCA to log transform value  , there are four principal cordinate component
print(ir.pca) #print  each of the four PCs, and their rotation (or loadings), which are the coefficients of the linear combinations of the continuous variables.
```

```r
plot(ir.pca, type='l')
```

```r
summary(ir.pca) #In this simple case with only 4 PCs this is not a hard task and we can see that the first two PCs explain most of the variability in the data
```

```r
iris.mat <- as.matrix(log.iris)
cov.mat <- cor(iris.mat)
eigen(cov.mat)  #eigne matrix by decomposition
```

```r
ggplot(df, aes(Petal.Length, Petal.Width)) + geom_point(aes(col=Species), size=4)
```

```r
set.seed(101)
irisCluster <- kmeans(df[,1:4], center=3, nstart=20) # three clusters  of  different sizes
irisCluster
```

```r
table(irisCluster$cluster, df$Species)
```

```r
library(cluster)
clusplot(iris, irisCluster$cluster, color=T, shade=T, labels=0, lines=0)
```

```r
tot.withinss <- vector(mode="character", length=10)
for (i in 1:10){
  irisCluster <- kmeans(df[,1:4], center=i, nstart=20)
  tot.withinss[i] <- irisCluster$tot.withinss
}
```

```r
plot(1:10, tot.withinss, type="b", pch=19)
```

```r
g <- ggbiplot(ir.pca, obs.scale = 1, var.scale = 1, 
              groups = iris.species, ellipse = TRUE, 
              circle = TRUE)
g <- g + scale_color_discrete(name = '')
g <- g + theme(legend.direction = 'horizontal', 
               legend.position = 'top')
print(g)
```
